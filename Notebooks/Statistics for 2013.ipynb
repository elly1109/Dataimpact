```python
## mport some important modules
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
from IPython.display import display
rcParams['figure.figsize'] = 15, 6
```

```python
## So, we can now load the dataset and look at some initial rows and data types of the columns:
data = pd.read_csv('../Data/IPD-Under-5-years-2013.csv')
print ('\n Data Types:')
data.dtypes
```

```python
## printing out some first few rows 
data.head()
```

```python
## Checking whether the dataset contains some missings values
data.isnull()
```

```python
### Alternative way
## This package can be used to display whether there is some missing package from your data
import missingno as msno
```

```python
msno.bar(df=data)
```

```python
## first 10 regions with large number of incidences
data.nlargest(10, data[['Total']], keep='first')
```

```python
## Regions with smallest number of incidences 
data.nsmallest(5, data[['Total']], keep='first')
```

```python
data.describe()
```

```python
### ftd means "first ten diseases with large occurance".
ftd=data.mean().sort_values(ascending=False)[0:11][1:]
ftd
```

```python
ftd.plot(kind="bar")
plt.title("Top ten disease in 2013 for \n under 5 IPD")
plt.ylabel("Number of inpatient department diagnoses (IPD)")
plt.savefig("../figure/2013_top_ten_disease.png", bbox_inches="tight", dpi=1200)
```
